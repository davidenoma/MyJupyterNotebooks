```python
import requests 
```

```python
# call the url from NCDC 
CovidNija = requests.get("https://covid19.ncdc.gov.ng/")
```

```python
print(CovidNija.text)
```

```python
# pass HTML using Beatutiful Soup 
from bs4 import BeautifulSoup 
soup = BeautifulSoup(CovidNija.text, "html.parser")
```

```python
tds = []
results = soup.find_all("table")
for tr in soup.find_all('tr')[1:]:
    tds.append(tr.find_all('td')) 
```

#### To seperate the records into four components 
###### States Affected 
###### No. of Cases (Lab Confirmed) 
###### No. of Cases (on admission) 
###### No. Discharged 
###### No. of Deaths 

```python
states = []
cases = []
admission = []
discharged = [] 
deaths = []
i = 0
while( len(tds) > i):
    items = [elem.getText() for elem in tds[i]]
    states.append(items[0].split('\n')[1])
    cases.append(items[1].split('\n')[0])
    admission.append(items[2].split('\n')[0])
    discharged.append(items[3].split('\n')[0])
    deaths.append(items[4].split('\n')[0])
    i += 1
    
print(states,cases,admission,discharged,deaths)
```
